```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://insights.blackcoffer.com/ai-in-healthcare-to-improve-patient-outcomes/"
```

```python
page01=requests.get(url)
#404- server can not be found
#204-no content in the actual web page
#400-server unable to find -bad request
```

```python
soup=BeautifulSoup(page01.text, 'html.parser')
```

```python
print(soup)
```

```python
print(soup.prettify)
```

```python
text = soup.get_text()
```

```python
print(text)
```

```python
article_title = soup.find("h1").get_text()

# extract title of the article.
```

```python
print(article_title)
```

```python
article_text = ""
paragraphs = soup.find_all("p")
for p in paragraphs:
    article_text += p.get_text() + "\n"

    # extracting the article text in the web
```

```python
print(article_text)
```

```python
pip install nltk
```

```python
nltk.download('all')
```

```python
# Sentiment analysis
#Sentimental analysis is the process of determining whether a piece of writing is positive, negative, or neutral
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
sentiment_scores = sia.polarity_scores(article_text)
positive_score = sentiment_scores['pos']
negative_score = sentiment_scores['neg']
polarity_score = sentiment_scores['compound']
subjectivity_score = sentiment_scores['neu'] + sentiment_scores['pos'] + sentiment_scores['neg']
```

```python
print(sentiment_scores)
print(negative_score)
print(polarity_score)
print(subjectivity_score)
```

```python
# Tokenization
#Tokenization is a crucial preprocessing step in natural language processing.
#Tokenization is the process of splitting a text into individual words or sentences, making it easier to analyze and process the text.

words = nltk.word_tokenize(article_text)

#This line uses NLTK's word_tokenize() function to tokenize the article_text into a list of individual words.
#Each word is represented as a separate element in the words list.

sentence_list = nltk.sent_tokenize(article_text)

#the words list will contain all the individual words from the article_text.
#And the sentence_list list will contain all the sentences from the article_text
```

```python
# Text analysis for the article given
avg_sentence_length = sum([len(sentence.split()) for sentence in sentence_list]) / len(sentence_list)
word_count = len(words)

avg_word_length = sum([len(word) for word in words]) / word_count
complex_word_count = len([word for word in words if len(word) > 6])

percentage_complex_words = (complex_word_count / word_count) * 100
fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)
```

```python
# Print the results
print("Article Title:", article_title)
print("Positive Score:", positive_score)
print("Negative Score:", negative_score)
print("Polarity Score:", polarity_score)
print("Subjectivity Score:", subjectivity_score)
print("Average Sentence Length:", avg_sentence_length)
print("Percentage of Complex Words:", percentage_complex_words)
print("FOG Index:", fog_index)
print("Average Word Length:", avg_word_length)
print("Complex Word Count:", complex_word_count)
print("Word Count:", word_count)
```
